```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# 1. Define the model
def lotka_volterra(y, t, alpha, beta, delta, gamma):
    prey, predator = y
    # Prey equation: dx/dt = alpha*x - beta*x*y
    dprey_dt = alpha * prey - beta * prey * predator
    # Predator equation: dy/dt = delta*beta*x*y - gamma*y
    dpredator_dt = delta * beta * prey * predator - gamma * predator
    return [dprey_dt, dpredator_dt]

# 2. Parameters
alpha = 1.0  # Prey growth rate
beta = 0.1   # Predation rate
delta = 0.1  # Predator reproduction rate per prey eaten
gamma = 0.5  # Predator mortality rate

# 3. Initial conditions and time grid
x0 = 40      # Initial prey population
y0 = 9       # Initial predator population
t = np.linspace(0, 50, 500) # 50 days, 500 points

# 4. Solve the ODE
solution = odeint(lotka_volterra, [x0, y0], t, args=(alpha, beta, delta, gamma))
prey, predator = solution.T

# 5. Plot the results
plt.figure(figsize=(10,5))
plt.plot(t, prey, label='Prey (Rabbits)')
plt.plot(t, predator, label='Predator (Foxes)')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('Lotka-Volterra Predator-Prey Model')
plt.legend()
plt.grid()
plt.show()

```
