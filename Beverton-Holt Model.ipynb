```python
import numpy as np
import matplotlib.pyplot as plt

def beverton_holt(n_t, R_0, K):
    """Single time step of Beverton-Holt model: n_{t+1} = R_0*n_t / (1 + n_t/K)"""
    return (R_0 * n_t) / (1 + n_t / K)

def beverton_holt_simulation(n_0, R_0, K, time_steps):
    """Simulate Beverton-Holt dynamics over time."""
    pop = np.zeros(time_steps + 1)
    pop[0] = n_0
    for t in range(time_steps):
        pop[t + 1] = beverton_holt(pop[t], R_0, K)
    return pop

# Parameters
n_0, R_0, K, time_steps = 5, 2, 100, 100
population = beverton_holt_simulation(n_0, R_0, K, time_steps)

# Plot
plt.figure(figsize=(9, 6))
plt.plot(population, 'b-o', markersize=4)
plt.axhline(K, color='r', linestyle='--', label=f'Carrying Capacity (K={K})')
plt.xlabel('Generation (t)', fontsize=12)
plt.ylabel('Population (n)', fontsize=12)
plt.title(f'Beverton-Holt Model: $R_0={R_0}$, $K={K}$, $n_0={n_0}$', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Results
print(f"Initial: {population[0]:.2f} | Final: {population[-1]:.2f} | Equilibrium: {K * (R_0 - 1) / R_0:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

R0 = 2.5
k = 100
n0 = 10
T = 50

n = np.zeros(T)
n[0] = n0

for t in range(T-1):
    n[t+1] = (R0 * n[t]) / (1 + n[t]/k)

plt.plot(n)
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Beverton-Holt Model")
plt.show()
```
