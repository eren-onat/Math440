## Dataset A: Smartphone Adoption in US

Percentage of US adults owning a smartphone (2011-2021)

```python
# Data
year_phone = np.array([2011, 2013, 2015, 2017, 2019, 2021])
ownership = np.array([35, 56, 68, 77, 81, 85])  # percentage

# Convert years to time since 2011 for modeling
t_phone = year_phone - 2011

# Plot the data
plt.figure(figsize=(8, 5))
plt.plot(year_phone, ownership, 'o', markersize=8, label='Data')
plt.xlabel('Year')
plt.ylabel('Smartphone Ownership (%)')
plt.title('US Smartphone Adoption')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("\nObserve the pattern: Is it linear? Exponential? Saturating?")
```

### Part (a): Identify and Fit Model

```python
# YOUR CODE HERE
# Based on the pattern, choose appropriate model:
# - Exponential growth: y(t) = y0 * exp(k*t)
# - Logistic growth: y(t) = K / (1 + A*exp(-r*t))
# - Power law: y(t) = a * t^b

# Define your chosen model function

def logistic_model(t, K, r, x0):
    """
    Standard logistic growth model: x(t) = K / (1 + A*exp(-r*t))
    where A = (K - x0) / x0
    
    Parameters:
    - K: carrying capacity (maximum sustainable population)
    - r: intrinsic growth rate (per unit time)
    - x0: initial population at t=0
    """
    A = (K - x0) / x0
    return K / (1 + A * np.exp(-r * t))

# Fit the model
# YOUR CODE HERE
# p0 = [...]

K_guess = 100 # Hint: max(t_phone) * 1.5  # Assume K is 50% above max observed
r_guess = 0.5 # YOUR CODE HERE  # Hint: 0.5               # Moderate growth rate
x0_guess = min(ownership) # YOUR CODE HERE  # Hint: x_bac[0]         # Use first data point

# params, cov = curve_fit(...)
# Lower bounds: K > max(data), r > 0, x0 > 0
# Upper bounds: K can be large, r < 10 (very fast growth), x0 < max(data)
bounds = (
    [max(ownership) * 1.01,0.01,0],      # Lower bounds
    [np.inf, 10, min(ownership)*1.01]      # Upper bounds
)

p0 = [K_guess, r_guess, x0_guess]

params_logistic, cov_logistic = curve_fit(
    logistic_model, t_phone, ownership, 
    p0=p0, 
    bounds=bounds,
    maxfev=10000  # Allow more iterations if needed
)

K_fit, r_fit, x0_fit = params_logistic
std_errors_log = np.sqrt(np.diag(cov_logistic))
# Report parameters
print("\nFitted Parameters:")
# YOUR CODE HERE

# Compute R²
x_pred = logistic_model(t_phone, K_fit, r_fit, x0_fit)

SS_res = np.sum((ownership - x_pred)**2)
SS_tot = np.sum((ownership - np.mean(ownership))**2)
R2 = 1 - SS_res / SS_tot

print(f"R^2 = {R2:.4f}")

print("\n=== Logistic Growth Fit ===")
print(f"Fitted parameters:")
print(f"  K (carrying capacity) = {K_fit:.1f} ± {std_errors_log[0]:.1f}")
print(f"  r (growth rate)       = {r_fit:.3f} ± {std_errors_log[1]:.3f} per hour")
print(f"  x0 (initial pop)      = {x0_fit:.1f} ± {std_errors_log[2]:.1f}")

# Compute and display inflection point
x_pred = logistic_model(t_phone, K_fit, r_fit, x0_fit)
residuals = ownership - x_pred

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

t_smooth =  np.linspace(0,20,100)
x_fit = logistic_model(t_smooth, K_fit, r_fit, x0_fit)
# Left: Data and fit
ax1.plot(t_phone, ownership, 'o', markersize=8, label='Data', color='steelblue')
ax1.plot(t_smooth, x_fit, '-', linewidth=2, label='Fitted model', color='darkorange')
ax1.set_xlabel('Time (hours)')
ax1.set_ylabel('Population')
ax1.set_title('Logistic Growth Fit')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: Residuals
ax2.plot(t_phone, residuals, 'o', markersize=8, color='darkred')
ax2.axhline(0, color='black', linestyle='--', linewidth=1)
ax2.set_xlabel('Time (hours)')
ax2.set_ylabel('Residual (Observed - Predicted)')
ax2.set_title('Residual Plot')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nResidual Analysis:")
print(f"Mean residual:     {np.mean(residuals):.2f} (should be ≈ 0)")
print(f"Std dev residual:  {np.std(residuals):.2f}")
print(f"Max |residual|:    {np.max(np.abs(residuals)):.2f}")
```

## Dataset B: Moore's Law - Transistor Count

Number of transistors on Intel processors (1971-2017)

```python
# Data
year_cpu = np.array([1971, 1978, 1985, 1993, 2000, 2008, 2017])
transistors = np.array([2.3, 29, 275, 3100, 42000, 731000, 19200000])  # thousands

# Convert to time since 1971
t_cpu = year_cpu - 1971

# Plot the data
plt.figure(figsize=(8, 5))
plt.plot(year_cpu, transistors, 'o', markersize=8, label='Data')
plt.xlabel('Year')
plt.ylabel('Transistor Count (thousands)')
plt.title("Moore's Law: Transistor Count Growth")
plt.yscale('log')  # Note: log scale on y-axis
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("\nObserve: On a log scale, does this look linear?")
```

### Part (a): Identify and Fit Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Define model
def transistor_model(t, A, k):
    return A * np.exp(k * t)

# Time since 1971
t = year_cpu - 1971

# Data (in thousands, as given)
y = transistors
logy = np.log(y)

# Fit log(y) = ln(A) + k t
k, lnA = np.polyfit(t, logy, 1)
A = np.exp(lnA)

# Predictions at observed points
y_hat = transistor_model(t, A, k)

# R^2 on original scale
SS_res = np.sum((y - y_hat)**2)
SS_tot = np.sum((y - np.mean(y))**2)
R2 = 1 - SS_res / SS_tot

print(f"A = {A:.4f}  (thousand transistors at t = 0, i.e., year 1971)")
print(f"k = {k:.4f}  (per year)")
print(f"R^2 = {R2:.4f}")

# Smooth curve for plotting
t_fine = np.linspace(0, t.max(), 200)

plt.figure(figsize=(8,5))
plt.plot(year_cpu, y, 'o', markersize=8, label='Data')
plt.plot(1971 + t_fine, transistor_model(t_fine, A, k), label='Exponential Fit')
plt.yscale('log')
plt.xlabel('Year')
plt.ylabel('Transistor Count (thousands)')
plt.title("Moore's Law: Exponential Fit")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### Part (b): Justification

Since the data was on a log scale I used an expoential model. Thus for fitting the model $y(t)=Ae^kt$ it was a high match giving a $R^2$ value of 0.97

## Dataset C: Coffee Cooling

Temperature of coffee cooling in a room at 22°C

```python
# Data
time_coffee = np.array([0, 2, 4, 6, 8, 10, 15, 20, 30])  # minutes
temp_coffee = np.array([95, 82, 71, 63, 56, 51, 42, 36, 28])  # degrees C
T_room = 22  # degrees C

# Plot
plt.figure(figsize=(8, 5))
plt.plot(time_coffee, temp_coffee, 'o', markersize=8, label='Data')
plt.axhline(T_room, color='red', linestyle='--', label='Room temperature')
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (°C)')
plt.title('Coffee Cooling')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print("\nNote: Temperature approaches room temperature asymptotically")
```

### Part (a): Identify and Fit Model

**Hint:** Newton's Law of Cooling states that the rate of temperature change is proportional to the difference from ambient temperature:

$$T(t) = T_{\text{room}} + (T_0 - T_{\text{room}})e^{-kt}$$

```python
# YOUR CODE HERE

def cooling_model(t, T_room, T0, k):
    """Newton's Law of Cooling"""
    return T_room + (T0 - T_room) * np.exp(-k * t)
T0_guess = temp_coffee[0]   # initial coffee temperature
k_guess = 0.1    
# Fit
# Note: T_room is known (22°C), but can also be fitted
# YOUR CODE HERE
params, cov = curve_fit(
    lambda t, T0, k: cooling_model(t, T_room, T0, k),
    time_coffee,
    temp_coffee,
    p0=[T0_guess, k_guess],
    maxfev=10000
)

T0_fit, k_fit = params
std_errors = np.sqrt(np.diag(cov))

T_pred = cooling_model(time_coffee, T_room, T0_fit, k_fit)

SS_res = np.sum((temp_coffee - T_pred)**2)
SS_tot = np.sum((temp_coffee - np.mean(temp_coffee))**2)
R2 = 1 - SS_res / SS_tot

# Report and compute R²
print("\nFitted Parameters:")
print(f"T0 = {T0_fit:.2f} ± {std_errors[0]:.2f} °C")
print(f"k  = {k_fit:.4f} ± {std_errors[1]:.4f} per minute")
print(f"R^2 = {R2:.4f}")
t_smooth = np.linspace(0, max(time_coffee), 200)
T_fit = cooling_model(t_smooth, T_room, T0_fit, k_fit)

plt.figure(figsize=(8,5))
plt.plot(time_coffee, temp_coffee, 'o', label='Data')
plt.plot(t_smooth, T_fit, '-', label='Cooling model')
plt.axhline(T_room, color='red', linestyle='--', label='Room temperature')
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (°C)')
plt.title("Newton's Law of Cooling Fit")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

### Part (b): Justification

The coffee temperature decreases rapidly at first and then more slowly, approaching room temperature as a slower fate. This behavior is similar to that of Newton’s Law of Cooling, which states that "the rate of temperature change is proportional to the difference between the object’s temperature and the ambient temperature". The fitted exponential decay model closely follows the data and produces a high $R^2$ value of 0.9969, indicating a strong fit.

# Problem 2: Tumor Growth - Competing Models

Compare exponential vs. Gompertz growth models for tumor volume data.

## Data

```python
# Tumor volume data
day_tumor = np.array([0, 10, 20, 30, 40, 50, 60])
volume_tumor = np.array([12, 28, 58, 105, 168, 238, 305])  # mm^3

# Plot
plt.figure(figsize=(8, 5))
plt.plot(day_tumor, volume_tumor, 'o', markersize=8, label='Data')
plt.xlabel('Day')
plt.ylabel('Tumor Volume (mm³)')
plt.title('Tumor Growth Data')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Part (a): Fit Both Models

```python
# Model A: Exponential
def exponential_model(t, V0, k):
    return V0 * np.exp(k * t)

# Fit exponential
p0_exp = [12, 0.05]

params_exp, cov_exp = curve_fit(
    exponential_model,
    day_tumor,
    volume_tumor,
    p0=p0_exp,
    maxfev=10000
)
V0_fit, k_fit = params_exp
std_errors_exp = np.sqrt(np.diag(cov_exp))
print("Exponential Model:")
print(f"  V0 = {V0_fit:.2f} ± {std_errors_exp[0]:.2f} mm^3")
print(f"  k  = {k_fit:.4f} ± {std_errors_exp[1]:.4f} per day")
# YOUR CODE HERE
V0_fit, k_fit = params_exp
std_errors_exp = np.sqrt(np.diag(cov_exp))
# Compute R² for exponential
V_pred_exp = exponential_model(day_tumor, V0_fit, k_fit)
SS_res = np.sum((volume_tumor - V_pred_exp)**2)
SS_tot = np.sum((volume_tumor - np.mean(volume_tumor))**2)
R2_exp = 1 - SS_res / SS_tot

print(f"  R^2 = {R2_exp:.4f}")
```

```python
# Model B: Gompertz
def gompertz_model(t, V0, a, b):
    """Gompertz: V(t) = V0 * exp(a * (1 - exp(-b*t)))"""
    return V0 * np.exp(a * (1 - np.exp(-b * t)))

# Fit Gompertz
# Initial guess: V0 ≈ 12, a ≈ 3-4, b ≈ 0.02-0.05
p0_gomp = [12, 3.5, 0.03]
params_gomp, cov_gomp = curve_fit(
    gompertz_model,
    day_tumor,
    volume_tumor,
    p0=p0_gomp,
    maxfev=10000
)
# Report
print("\nGompertz Model:")
V0_gomp, a_gomp, b_gomp = params_gomp
std_errors_gomp = np.sqrt(np.diag(cov_gomp))

print("\nGompertz Model:")
print(f"  V0 = {V0_gomp:.2f} ± {std_errors_gomp[0]:.2f} mm^3")
print(f"  a  = {a_gomp:.3f} ± {std_errors_gomp[1]:.3f}")
print(f"  b  = {b_gomp:.4f} ± {std_errors_gomp[2]:.4f} per day")

V_pred_gomp = gompertz_model(day_tumor, V0_gomp, a_gomp, b_gomp)

# Compute R² for Gompertz
SS_res = np.sum((volume_tumor - V_pred_gomp)**2)
SS_tot = np.sum((volume_tumor - np.mean(volume_tumor))**2)
R2_gomp = 1 - SS_res / SS_tot

print(f"  R^2 = {R2_gomp:.4f}")
```

## Part (b): Comparison Figure

```python
# Create two-panel figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left panel: Both fits
t_smooth = np.linspace(0, 60, 200)
# Compute predictions for both models
V_exp_smooth = exponential_model(t_smooth, V0_fit, k_fit)
V_gomp_smooth = gompertz_model(t_smooth, V0_gomp, a_gomp, b_gomp)
ax1.plot(day_tumor, volume_tumor, 'o', markersize=8, label='Data', color='steelblue')
ax1.plot(t_smooth, V_exp_smooth, '-', label='Exponential', color='orange')
ax1.plot(t_smooth, V_gomp_smooth, '--', label='Gompertz', color='green')
ax1.set_xlabel('Day')
ax1.set_ylabel('Volume (mm³)')
ax1.set_title('Model Comparison')
ax1.legend()
ax1.grid(True, alpha=0.3)



# Compute residuals for both 
residuals_exp = volume_tumor - exponential_model(day_tumor, V0_fit, k_fit)
residuals_gomp = volume_tumor - gompertz_model(day_tumor, V0_gomp, a_gomp, b_gomp)


ax2.plot(day_tumor, residuals_exp, 'o', markersize=8, label='Exponential', color='orange')
ax2.plot(day_tumor, residuals_gomp, 's', markersize=8, label='Gompertz', color='green')
ax2.axhline(0, color='black', linestyle='--', linewidth=1)
ax2.set_xlabel('Day')
ax2.set_ylabel('Residual (mm³)')
ax2.set_title('Residual Comparison')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Part (c): Model Comparison Using AIC

The Akaike Information Criterion (AIC) balances fit quality with model complexity:

$$\text{AIC} = n\ln\left(\frac{\text{SSE}}{n}\right) + 2p$$

where $n$ = number of data points, $p$ = number of parameters.

**Lower AIC is better.** Difference > 2 is significant.

```python
# Compute AIC for exponential model
n = len(day_tumor)
p_exp = 2  # Number of parameters (V0, k)
SSE_exp = np.sum((volume_tumor - exponential_model(day_tumor, V0_fit, k_fit))**2)
AIC_exp = n * np.log(SSE_exp / n) + 2 * p_exp

# Compute AIC for Gompertz model
p_gomp = 3  # Number of parameters (V0, a, b)
SSE_gomp = np.sum((volume_tumor - gompertz_model(day_tumor, V0_gomp, a_gomp, b_gomp))**2)
AIC_gomp = n * np.log(SSE_gomp / n) + 2 * p_gomp

print("Model Comparison:")
print(f"Exponential: AIC = {AIC_exp:.2f}, SSE = {SSE_exp:.2f}, R² = {R2_exp:.4f}")
print(f"Gompertz:    AIC = {AIC_gomp:.2f}, SSE = {SSE_gomp:.2f}, R² = {R2_gomp:.4f}")
print(f"\nΔAIC = {abs(AIC_exp - AIC_gomp):.2f}")

if AIC_gomp < AIC_exp:
    print("→ Gompertz model is preferred (lower AIC)")
else:
    print("→ Exponential model is preferred (lower AIC)")
```

# Problem 3: Allometric Scaling with Prediction

Metabolic rate scales with body mass: $M = am^b$

## Data

```python
# Allometric data
animals = ['Mouse', 'Rat', 'Rabbit', 'Dog', 'Human', 'Cow', 'Elephant']
mass = np.array([0.02, 0.3, 3, 15, 70, 400, 4000])  # kg
metabolic_rate = np.array([0.05, 0.5, 3.5, 12, 80, 300, 1800])  # watts

print("Animal       | Mass (kg) | Metabolic Rate (W)")
print("-" * 50)
for animal, m, M in zip(animals, mass, metabolic_rate):
    print(f"{animal:12s} | {m:9.2f} | {M:18.1f}")
```

## Part (a): Linearize and Fit

```python
# Transform to log scale
X = np.log(mass)
Y = np.log(metabolic_rate)

# Linear regression
b, intercept = np.polyfit(X, Y, 1)

# Extract parameters
a = np.exp(intercept)

print(f"Fitted Power Law: M = {a:.3f} * m^{b:.3f}")
print(f"\nIs b close to Kleiber's prediction of 0.75? Yes-ish!")
# Log-log plot
plt.figure(figsize=(8, 6))
plt.plot(X, Y, 'o', label='Data')
# fitted line in log-log space
X_line = np.linspace(X.min(), X.max(), 200)
Y_line = intercept + b * X_line
plt.plot(X_line, Y_line, '-', label='Fitted line')

plt.xlabel('ln(Mass [kg])')
plt.ylabel('ln(Metabolic Rate [W])')
plt.title('Allometric Scaling (Log-Log Plot)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Part (b): Prediction with Uncertainty

Predict metabolic rate for a **capybara** (mass = 50 kg) with 95% prediction interval.

**Formula for prediction interval (in log-space):**

$$\text{CI} = \hat{y} \pm t_{n-2, 0.025} \cdot s \sqrt{1 + \frac{1}{n} + \frac{(x_{\text{new}} - \bar{x})^2}{\sum(x_i - \bar{x})^2}}$$

where $s = \sqrt{\frac{\text{SSE}}{n-2}}$ and $t_{n-2, 0.025} \approx 2.57$ for $n=7$.

```python
# Capybara mass
m_capybara = 50  # kg
x_new = np.log(m_capybara)

# Prediction in log-space
# YOUR CODE HERE
y_pred_log = intercept + b * x_new

# Compute standard error
Y_pred_all = intercept + b * X
residuals = Y - Y_pred_all
SSE = np.sum(residuals**2)
s = np.sqrt(SSE / (n - 2))

# Prediction interval
n = len(X)
t_crit = 2.57  # t_{5, 0.025}
x_mean = np.mean(X)
# YOUR CODE HERE
margin = t_crit * s * np.sqrt(1 + 1/n + (x_new - x_mean)**2 / np.sum((X - x_mean)**2))

# Confidence interval in log-space
# YOUR CODE HERE
CI_lower_log = y_pred_log - margin
CI_upper_log = y_pred_log + margin

# Convert back to original scale
M_pred = np.exp(y_pred_log)
CI_lower = np.exp(CI_lower_log)
CI_upper = np.exp(CI_upper_log)

print(f"Predicted metabolic rate for capybara (50 kg):")
print(f"  Point estimate: {M_pred:.1f} W")
print(f"  95% PI: [{CI_lower:.1f}, {CI_upper:.1f}] W")
print(f"\nActual measured value: 48 W")
print(f"Does 48 W fall within the prediction interval? Yes!")
```

**Interpretation:**

Yes the model falls under the actual model which means the model is very good
